```python

top_10_horizontal = gb.sort_values(by='NET REVENUE', ascending=True).tail(10)


ax = top_10_horizontal.plot(kind='barh', figsize=(12, 6), legend=False)


for container in ax.containers:
    ax.bar_label(container, fmt=lambda x: f' ${x/1e6:.2f}M', padding=3)

plt.title('Top 10 Stores by Net Revenue', fontsize=14, pad=15)
plt.xlabel('Net Revenue (USD)')
plt.tight_layout()
plt.show()
```

```python

cohort_counts = df.groupby('EBITDA COHORT')['STORE'].nunique().reset_index(name='STORE COUNT')


logical_order = ['Less than 0%', '0% to 10%', '10% to 20%', 'More than 20%']


cohort_counts['EBITDA COHORT'] = pd.Categorical(
    cohort_counts['EBITDA COHORT'], 
    categories=logical_order, 
    ordered=True
)


cohort_counts = cohort_counts.sort_values(by='EBITDA COHORT') 



ax = cohort_counts.plot(
    x='EBITDA COHORT', 
    y='STORE COUNT', 
    kind='bar', 
    figsize=(10, 5), 
    legend=False,
    color='skyblue'
)


for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3)


plt.title('Number of Unique Stores by EBITDA COHORT Bucket', fontsize=14, pad=15)
plt.ylabel('Unique Store Count')
plt.xlabel('EBITDA Cohort')
plt.xticks(rotation=0) 
plt.tight_layout()
plt.show()
```

```python
def variance_bucket(x):

    if 10000<x<17000:
        return 'Low'

    elif 17000<=x<23000:
        return 'Medium'
        
    else:
        return 'High'
```

```python
df['VARIANCE_BUCKET'] = df['VARIANCE'].apply(variance_bucket)
```

```python
df.sample(5)
```

```python
variance_summary = (
    df
    .groupby(
        ['REVENUE COHORT', 'VARIANCE_BUCKET','MONTH']
    )['STORE']
    .nunique()
    .reset_index(name='STORE_COUNT')
)

variance_summary
```

```python
variance_summary.head()
```

```python
total_stores = (
    variance_summary
    .groupby(['REVENUE COHORT','MONTH'])['STORE_COUNT']
    .sum()
    .reset_index(name='TOTAL_STORES')
)

total_stores
```

```python
variance_summary = variance_summary.merge(
    total_stores,
    on= ['REVENUE COHORT','MONTH']
)

variance_summary
```

```python
variance_summary['STORE_PERCENT'] = (
    variance_summary['STORE_COUNT']
    /
    variance_summary['TOTAL_STORES']
) * 100

variance_summary
```

```python
variance_summary.to_csv('variance_summary.csv', index=False)
```
